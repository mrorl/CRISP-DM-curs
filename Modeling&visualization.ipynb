## Этап 1: геокодирование.
Сначала нам потребуется использовать библиотеку geopy.geocoders для того, чтобы по адресу объекта найти его географические координаты и иметь возможность отобразить его на карте.



```python
from geopy.geocoders import Nominatim
```

## Геокодирование для df_cameras_adress.csv

```python
data = pd.read_csv("df_cameras_adress.csv")
```

```python
data
```

```python
data['address'] = data['city'] + ', ' + data['street'] + ', ' + data['house']
data.head()
```

```python
geolocator = Nominatim()

data['location'] = np.nan
data['latitude'] = np.nan
data['longitude'] = np.nan

for row, addr in enumerate(data['address']):
  location = geolocator.geocode(addr)
  if location is not None:
    print(row, location)
    data['location'][row] = location.address
    data['latitude'][row] = location.latitude
    data['longitude'][row] = location.longitude
```

```python
data.head()
```

```python
data.info()
```

Видим, что с помощью библиотеки удалось геокодировать не все адреса. Удалим записи с такими адресами.

```python
data =data.dropna()
```

```python
data.info()
```

Сохраним датасет с полученными координатами объектов.

```python
data.to_csv('final_df_cameras.csv', index = False, header=True)
```

# Моделирование
Задачи:
- с помощью полученного совмещенного датасета камер и зданий под аренду сделать предсказание относительно количества камер в зависимости от района города, площади и типа помещения;
- рассчитать возможную прибыль от установки видеокамер.

```python
with_cameras = pd.read_csv("training_data.csv")
with_no_cameras = pd.read_csv("predict_data.csv")
```

Простыми словами, мы высчитываем, сколько в других районах, на примере идеального, в теории можно будет добавить камер.

В качестве "эталонных" были взяты Адмиралтейский и Центральный районы. В этих двух районах, благодаря визуализации, хорошо видно, что плотность покрытия сдаваемой площади камерами значительно больше.

##Адмиралтейский район##

```python
adm_district = with_cameras.loc[with_cameras['district'] == 'Адмиралтейский']
adm_district_no_cameras = with_no_cameras.loc[with_no_cameras['district'] == 'Адмиралтейский']
```

```python
area_without_cams = adm_district_no_cameras['area'].sum()
area_with_cams = adm_district['area'].sum()

total_area = area_with_cams + area_without_cams
total_cams = adm_district['amount'].sum()
meter_cam_ratio_adm = total_area/total_cams

print('Сдаваемая площадь в Адмиралтейском р-не: ', total_area)
print('Число камер на фасадах сдаваемых помещений Адмиралтейского р-на: ', total_cams)
print('Сотношение сдаваемой площади к количеству камер в Адмиралтейском р-не: ', meter_cam_ratio_adm)
```

##Центральный район##

```python
cnt_district = with_cameras.loc[with_cameras['district'] == 'Центральный']
cnt_district_no_cameras = with_no_cameras.loc[with_no_cameras['district'] == 'Центральный']
```

```python
cnt_area_without_cams = cnt_district_no_cameras['area'].sum()
cnt_area_with_cams = cnt_district['area'].sum()

total_area = cnt_area_with_cams + cnt_area_without_cams
total_cams = cnt_district['amount'].sum()
meter_cam_ratio_cnt = total_area/total_cams

print('Сдаваемая площадь в Центральном р-не: ', total_area)
print('Число камер на фасадах сдаваемых помещений Центрального р-на: ', total_cams)
print('Сотношение сдаваемой площади к количеству камер в Центральном р-не: ', meter_cam_ratio_cnt)
```

Возьмем среднее значение коэффициента, по которому отсеим районы с лучшей плотностью покрытия нежели среднее значение эталонных районов. Это позволит нам убрать районы в которых мало сдаваемой площади (такие районы заведомо считаем менее прибыльными).

```python
mean_meter_cam_ration = (meter_cam_ratio_adm + meter_cam_ratio_cnt)/2
mean_meter_cam_ration
```

```python
data_by_districts = []
unique_districts = with_cameras.district.unique()
for district in unique_districts:
  filtered_district_cameras = with_cameras.loc[with_cameras['district'] == district]
  filtered_district_no_cameras = with_no_cameras.loc[with_no_cameras['district'] == district]

  tmp_area_without_cams = filtered_district_no_cameras['area'].sum()
  tmp_area_with_cams = filtered_district_cameras['area'].sum()

  tmp_total_area = tmp_area_without_cams + tmp_area_without_cams
  tmp_total_cams = filtered_district_cameras['amount'].sum()

  tmp_meter_cam_ratio_adm = tmp_total_area/tmp_total_cams
  
  if tmp_meter_cam_ratio_adm > mean_meter_cam_ration:
      data_by_districts.append({'district': district, 'total_area': tmp_total_area, 
                                'total_cams': tmp_total_cams, 'ratio': tmp_meter_cam_ratio_adm})

data_by_districts
```

##Рейтинг прибыльности районов##


Выведем рейтинг прибыльных районов. Посчитаем для них максимально возможную выручку за единоразовую установку без учета стоимости камеры (4000₽ за камеру) и максимально возможную месячную прибыль за обслуживание камер района (280₽ за камеру в месяц).

```python
profit_by_district = []
for item in data_by_districts:
  new_ratio = item['ratio'] / mean_meter_cam_ration
  possible_cams_added = ((item['total_cams'] * new_ratio) - item['total_cams']).round()
  profit_by_district.append({'district': item['district'], 'possible_cams_added': possible_cams_added, })

profit_by_district = reversed(sorted(profit_by_district, key=lambda k: k['possible_cams_added']))

for item in profit_by_district:
  print(item['district'])
  cam_cost = 4000
  cam_maintenance_const = 280
  print("Максимальная спрос на камеры в районе (шт.): ", item['possible_cams_added'])
  print("Максимальная прибыль от установки (₽): ", item['possible_cams_added'] * cam_cost)
  print("Максимальная прибыль от обслуживания (₽/мес.): ", item['possible_cams_added'] * cam_maintenance_const)
  print('\n')
```
